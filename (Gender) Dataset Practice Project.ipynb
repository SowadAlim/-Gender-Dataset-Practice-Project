```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd


df=pd.read_csv("survey_results_public.csv")

df
```

# Singling out the "Gender" Column

```python
df1= df[["Gender"]]
df1.head(100)
```

# Dropping the "N/A"

```python
df2 = df1.dropna()
df2
```

```python
df2 = df2.reset_index(drop = True)
df2
```
